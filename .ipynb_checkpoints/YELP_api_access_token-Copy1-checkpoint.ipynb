# Getting the list open restaurants keywords

```python
keywords = []
for i in range(2000):
    url = '{0}{1}'.format(API_HOST, quote(SEARCH_PATH.encode('utf8')))
    headers = {'Authorization': 'Bearer %s' % bearer_token,}
    location = random.choice(nh)+', chicago'
    url_params = {
        'term': 'restaurants',
        'location': location.replace(' ', '+'),
        'limit': 50,
        'offset': random.choice(list(np.arange(0, 1000, 50))),
        'sort_by': random.choice(['best_match', 'rating', 'review_count', 'distance'])
        }        
    try:
        restaurants = requests.request('GET', url, headers=headers, params=url_params).json()
        if len(restaurants['businesses']) == 0:
            url_params['offset'] = random.choice(list(np.arange(0, restaurants['total'], 50)))
            restaurants = requests.request('GET', url, headers=headers, params=url_params).json()
        keyword = random.choice(restaurants['businesses'])['id']
        if keyword not in keywords:
            keywords.append(keyword)
        print(i, keyword)
    except:
        pass
```
