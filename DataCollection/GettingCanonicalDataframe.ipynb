```python
import pandas as pd
import numpy as np
from os import listdir
from os.path import isfile, join
```

```python
def CombineInfoReview(infocsv=None, reviewcsv=None):
    '''Replicate info of a given restaurant and combine it with its reviews into a dataframe'''
    df_Info = pd.read_csv(infocsv)    
    df_Review = pd.read_csv(reviewcsv).drop(['text'],axis=1)
    try:
        df_Info['first_review'] = [df_Review['date'].iloc[-1]]
        df_Info['last_review'] = [df_Review['date'].iloc[0]]
    except:
        df_Info['first_review'] = ['MISSING']
        df_Info['last_review'] = ['MISSING']        
    df_Info_copy = df_Info.copy()        
    for i in range(df_Review.shape[0]-1):
        df_Info = df_Info.append(df_Info_copy, ignore_index=True)
    df = df_Info.join(df_Review)
    return df
```

```python
def Getdf_InfoReview(path='./Gayot/'):
    '''Get info and reviews of restaurants inside a directory'''
    files = [f for f in listdir(path)]
    Reviewfiles = [f for f in files if '_review.csv' in f]
    Infofiles = [f[:-11]+'.csv' for f in Reviewfiles if f[:-11]+'.csv' in files]
    Info = path + Infofiles[0]
    Review = Info[:-4] + '_review.csv'
    df = CombineInfoReview(infocsv=Info, reviewcsv=Review)
    df['id'] = Infofiles[0][:-4]
    for i, f in enumerate(Infofiles[1:]):
        print(i+1, f)
        Info = path + f
        Review = Info[:-4] + '_review.csv'
        newdf = CombineInfoReview(infocsv=Info, reviewcsv=Review)
        newdf['id'] = f[:-4]
        df = df.append(newdf, ignore_index=True)            
    return df
```
